Q1.

A for loop is used when we know the number of iterations required in advance, and a while loop is used when we don't know the number of iterations in advance, or when we need to iterate until a specific condition is met.

Example for a for loop:

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)
This loop will iterate over the fruits list and print each fruit.

Example for a while loop:
num = 1
while num <= 10:
    print(num)
    num += 1
This loop will print the numbers from 1 to 10 because we don't know how many times we need to print the numbers in advance.

Q4.Using a for loop:

numbers = list(range(1, 101))
cubes = []
for number in numbers:
    cube = number ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes.append(number)
print(cubes)
Using a while loop:

numbers = list(range(1, 101))
cubes = []
i = 0
while i < len(numbers):
    cube = numbers[i] ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes.append(numbers[i])
    i += 1
print(cubes)
